```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/Amazon Sale Report.csv', encoding='latin-1') # or 'cp1252', 'ISO-8859-1'
```

```python
df.head()
```

```python
df.info()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='most_frequent')
df[['currency','Amount','ship-city','ship-state','ship-country','fulfilled-by','ship-postal-code']] = imputer.fit_transform(df[['currency','Amount','ship-city','ship-state','ship-country','fulfilled-by','ship-postal-code']])
```

```python
df.head()
```

```python
df = df.drop(['index','New','PendingS'], axis = 1)
```

```python
df
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt

# 1️⃣ Total revenue
total_revenue = df['Amount'].sum()

# 2️⃣ Average Order Value (AOV)
aov = df['Amount'].mean()

# 3️⃣ Total Quantity Sold
total_qty = df['Qty'].sum()

# 4️⃣ Monthly Sales Trends
df['Date'] = pd.to_datetime(df['Date'])  # This line is crucial

df['Year-Month'] = df['Date'].dt.to_period('M')
monthly_sales = df.groupby('Year-Month')['Amount'].sum()



# Plot Monthly Sales
plt.figure(figsize=(12, 5))
monthly_sales.plot(kind='line', marker='o', color='b')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Print key metrics
print(f"Total Revenue: ₹{total_revenue:,.2f}")
print(f"Average Order Value: ₹{aov:,.2f}")
print(f"Total Quantity Sold: {total_qty:,}")
```

```python
import seaborn as sns

# 1️⃣ Best-Selling Categories (by quantity)
top_categories_qty = df.groupby('Category')['Qty'].sum().sort_values(ascending=False)

# 2️⃣ Most Profitable Categories (by revenue)
top_categories_revenue = df.groupby('Category')['Amount'].sum().sort_values(ascending=False)

# 3️⃣ Average Order Quantity per Category
avg_qty_per_category = df.groupby('Category')['Qty'].mean().sort_values(ascending=False)

# Plot Best-Selling Categories (by Quantity)
plt.figure(figsize=(12, 5))
sns.barplot(x=top_categories_qty.index, y=top_categories_qty.values, palette="viridis")
plt.title("Top Best-Selling Categories (by Quantity)")
plt.xlabel("Category")
plt.ylabel("Total Quantity Sold")
plt.xticks(rotation=45)
plt.show()

# Plot Most Profitable Categories (by Revenue)
plt.figure(figsize=(12, 5))
sns.barplot(x=top_categories_revenue.index, y=top_categories_revenue.values, palette="magma")
plt.title("Top Most Profitable Categories (by Revenue)")
plt.xlabel("Category")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.show()

# Print key insights
print("🔹 Best-Selling Categories by Quantity:\n", top_categories_qty)
print("\n🔹 Most Profitable Categories by Revenue:\n", top_categories_revenue)
print("\n🔹 Average Order Quantity per Category:\n", avg_qty_per_category)
```

```python
# 1️⃣ Top Categories by State (Total Quantity)
top_categories_state = df.groupby(['ship-state', 'Category'])['Qty'].sum().reset_index()
top_categories_state = top_categories_state.sort_values(['ship-state', 'Qty'], ascending=[True, False])

# 2️⃣ Top Categories by City (Total Quantity)
top_categories_city = df.groupby(['ship-city', 'Category'])['Qty'].sum().reset_index()
top_categories_city = top_categories_city.sort_values(['ship-city', 'Qty'], ascending=[True, False])

# 3️⃣ Category-wise Revenue by State
category_revenue_state = df.groupby(['ship-state', 'Category'])['Amount'].sum().reset_index()
category_revenue_state = category_revenue_state.sort_values(['ship-state', 'Amount'], ascending=[True, False])

# Plot: Top Categories in States
plt.figure(figsize=(14, 6))
sns.barplot(x=top_categories_state['ship-state'], y=top_categories_state['Qty'], hue=top_categories_state['Category'], dodge=False)
plt.title("Top Selling Categories in Each State")
plt.xlabel("State")
plt.ylabel("Total Quantity Sold")
plt.xticks(rotation=90)
plt.legend(title="Category", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Plot: Category-wise Revenue Across States
plt.figure(figsize=(14, 6))
sns.barplot(x=category_revenue_state['ship-state'], y=category_revenue_state['Amount'], hue=category_revenue_state['Category'], dodge=False)
plt.title("Category-Wise Revenue in Each State")
plt.xlabel("State")
plt.ylabel("Total Revenue")
plt.xticks(rotation=90)
plt.legend(title="Category", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Print key insights
print("🔹 Top Categories Sold in Each State:\n", top_categories_state.head(20))
print("\n🔹 Top Categories Sold in Each City:\n", top_categories_city.head(20))
print("\n🔹 Category-Wise Revenue by State:\n", category_revenue_state.head(20))
```

```python
# Mapping order statuses to sentiment
def get_sentiment(status):
    if status in ["Shipped", "Shipped - Delivered to Buyer",]:
        return "Positive"
    elif status in ["Shipped - Returned to Seller", "Shipped - Returning to Seller", "Cancelled","Shipped - Rejected by Buyer","Shipped - Lost in Transit"]:
        return "Negative"
    else:
        return "Neutral"

df["Sentiment"] = df["Status"].apply(get_sentiment)

# Count sentiment categories
sentiment_counts = df["Sentiment"].value_counts()

# Plot sentiment distribution
plt.figure(figsize=(8, 5))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, palette=["green", "gray", "red"])
plt.title("Customer Sentiment Distribution (Based on Order Status)")
plt.xlabel("Sentiment")
plt.ylabel("Number of Orders")
plt.show()

# Print summary
print("🔹 Sentiment Analysis Based on Order Status:\n", sentiment_counts)
```

```python
# Count refund-related orders
refund_statuses = ["Refunded", "Returned", "Cancelled"]
refund_counts = df[df["Status"].isin(refund_statuses)].groupby("Category")["Order ID"].count().sort_values(ascending=False)

# Count 'Lost in Transit' cases
lost_orders = df[df["Status"] == "Shipped - Lost in Transit"].groupby("Category")["Order ID"].count()

# Plot refunds per category
plt.figure(figsize=(12, 6))
sns.barplot(x=refund_counts.index, y=refund_counts.values, palette="magma")
plt.title("Total Refunds by Category")
plt.xlabel("Category")
plt.ylabel("Number of Refunds")
plt.xticks(rotation=45)
plt.show()

print("🔹 Number of Refunds by Category:\n", refund_counts.head(10))
print("\n🔹 Lost in Transit Cases by Category:\n", lost_orders.head(10))
```

```python
# Define statuses that cause a loss
loss_statuses = ["Shipped - Returned to Seller", "Shipped - Returning to Seller", "Cancelled","Shipped - Rejected by Buyer","Shipped - Lost in Transit"]

# Create a new column 'Loss Amount'
df["Loss Amount"] = df["Amount"].where(df["Status"].isin(loss_statuses), 0)

# Display total losses
total_loss = df["Loss Amount"].sum()
print(f"💰 Total Loss Amount: ₹{total_loss:,.2f}")

# Show top categories with highest losses
loss_by_category = df.groupby("Category")["Loss Amount"].sum().sort_values(ascending=False)
print("\n🔹 Losses by Category:\n", loss_by_category.head(10))

# Plot losses per category
plt.figure(figsize=(12, 6))
sns.barplot(x=loss_by_category.index, y=loss_by_category.values, palette="Reds_r")
plt.title("Total Loss Amount by Category")
plt.xlabel("Category")
plt.ylabel("Loss Amount (₹)")
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count orders in each status
status_counts = df['Status'].value_counts()

# Plot bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=status_counts.index, y=status_counts.values, palette='viridis')
plt.xlabel("Order Status")
plt.ylabel("Number of Orders")
plt.title("Order Status Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
# Count orders with status "Shipped" or "Shipped - Delivered to Buyer"
fulfilled = df[(df['Status'] == "Shipped") | (df['Status'] == "Shipped - Delivered to Buyer")].shape[0]

# Count orders with other statuses (unfulfilled)
unfulfilled = df[~((df['Status'] == "Shipped") | (df['Status'] == "Shipped - Delivered to Buyer"))].shape[0]

labels = ['Fulfilled Orders', 'Unfulfilled Orders']
sizes = [fulfilled, unfulfilled]
colors = ['green', 'red']

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90, wedgeprops={'edgecolor': 'black'})
plt.title("Fulfillment Rate")
plt.show()
```
